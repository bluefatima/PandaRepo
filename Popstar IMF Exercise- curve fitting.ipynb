```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#Kroupa_2001 IMF. cluster mass of 10**4 M_sol
mass_array = np.loadtxt("masses.txt", delimiter = " " )

def f(x, a, b):
     return a*x**b

heights, bins = np.histogram(mass_array, bins = np.logspace(-3,3,50))

#dumb name- midpoints is really the same as bins, just truncating the last value to make it the same length as heights.
midpoints = bins[:-1]

midpoints1 = [i for i in midpoints if i < 8*10**-2]    
midpoints2 = [i for i in midpoints if (i >= 8*10**-2 and i < 5*10**-1)]
midpoints3 = [i for i in midpoints if i > 5*10**-1]

heights1, heights2, heights3 = np.split(heights, [len(midpoints1), len(midpoints1) + len(midpoints2)])

popt1, pcov1 = curve_fit(f, midpoints1, heights1, [10.0**4, 0.7])
popt2, pcov2 = curve_fit(f, midpoints2, heights2, [10.0**3, -0.3])
popt3, pcov3 = curve_fit(f, midpoints3, heights3, [10.0**2.7, -1.3])

#plt.hist(mass_array, bins = np.logspace(-3,3,50))
plt.loglog(midpoints,heights)
plt.plot(np.logspace(-2,-1,50), f(np.logspace(-2,np.log10(8*10**-2)), *popt1))
plt.plot(np.logspace(-1,0,50), f(np.logspace(np.log10(8*10**-2),np.log10(5*10**-1),50), *popt2))
plt.plot(np.logspace(0,2,50), f(np.logspace(np.log10(5*10**-1),2,50), *popt3))
plt.gca().set_xscale("log")
plt.gca().set_yscale("log")
plt.show()
```

```python
#Kroupa_2001 IMF. cluster mass of 10**4 M_sol
mass_array = np.loadtxt("masses.txt", delimiter = " " )

heights, binsz = np.histogram(mass_array, bins = np.logspace(-3,3,50))

xvals1 = np.arange(0.01,0.08,0.001)
yvals1 = 1.15923 * 10**4 * xvals1**0.7
xvals2 = np.arange(0.08,0.5, 0.01)
yvals2 = 9.27387 * 10**2 * xvals2**-0.3
xvals3 = np.arange(0.5,100,1)
yvals3 = 4.63693 * 10**2 * xvals3**-1.3
plt.plot(xvals1, yvals1,linewidth=2.0)
plt.plot(xvals2, yvals2,linewidth=2.0)
plt.plot(xvals3, yvals3,linewidth=2.0)
plt.hist(mass_array, bins = np.logspace(-3,3,50))
plt.xscale('log')
plt.yscale('log')
plt.show()

plt.loglog(binsz[:-1], heights, drawstyle ='steps-pre')
plt.show()

plt.hist(mass_array, bins = binsz)
plt.xscale('log')
plt.yscale('log')
plt.show()
```
