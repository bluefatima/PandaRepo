The goal here is to gain familiarity with Popstar and making an IMF. Here we make a cluster of 10^4, 10^5, and 10^6 solar masses using the Kroupa 2001 IMF, and overplot the analytical power law distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#Kroupa_2001 IMF. cluster mass of 10**4 M_sol
mass_array = np.loadtxt("masses.txt", delimiter = " " )
print(len(mass_array))
print(mass_array[10])

xvals1 = np.arange(0.01,0.08,0.001)
yvals1 = xvals1**0.7
xvals2 = np.arange(0.08,0.5, 0.01)
yvals2 = xvals2**-0.3
xvals3 = np.arange(0.5,100,1)
yvals3 = xvals3**-1.3
plt.plot(xvals1, 10**4 * yvals1,linewidth=2.0)
plt.plot(xvals2, 10**3 * yvals2,linewidth=2.0)
plt.plot(xvals3, 10**2.7 * yvals3,linewidth=2.0)
plt.hist(mass_array, bins = np.logspace(-3,3,50))
plt.gca().set_xscale("log")
plt.gca().set_yscale("log")
plt.show()
```

```python
#Kroupa_2001 IMF. cluster mass of 10**5
mass_array2 = np.loadtxt("masses2.txt", delimiter = " ")
print(len(mass_array2))
print(mass_array2[10])
```

```python
xvals1 = np.arange(0.01,0.08,0.005)
yvals1 = xvals1**0.7
xvals2 = np.arange(0.08,0.5, 0.05)
yvals2 = xvals2**-0.3
xvals3 = np.arange(0.5,500,10)
yvals3 = xvals3**-1.3
plt.plot(xvals1, 10**5 * yvals1,linewidth=2.0)
plt.plot(xvals2, 10**4 * yvals2,linewidth=2.0)
plt.plot(xvals3, 10**3.7 * yvals3,linewidth=2.0)
plt.hist(mass_array2, bins = np.logspace(-2,3,50))
plt.gca().set_xscale("log")
plt.gca().set_yscale("log")
plt.show()
```

```python
#Kroupa_2001 IMF. cluster mass of 10**6
mass_array3 = np.loadtxt("masses3.txt", delimiter = " ")
print(len(mass_array3))
print(mass_array3[10])
```

```python
xvals1 = np.arange(0.01,0.08,0.005)
yvals1 = xvals1**0.7
xvals2 = np.arange(0.08,0.5, 0.05)
yvals2 = xvals2**-0.3
xvals3 = np.arange(0.5,10000,10)
yvals3 = xvals3**-1.3
plt.plot(xvals1, 10**6 * yvals1,linewidth=2.0)
plt.plot(xvals2, 10**5 * yvals2,linewidth=2.0)
plt.plot(xvals3, 10**4.7* yvals3,linewidth=2.0)
plt.hist(mass_array3, bins = np.logspace(-2,5,50))
plt.gca().set_xscale("log")
plt.gca().set_yscale("log")
plt.show()
```

General sanity check: the length of the mass array tells us how many stars are in the cluster. There are about 10 times as many stars in the 10^5 solar mass cluster than the 10^4 solar mass cluster, which makes sense. Same deal for the 10^5 and 10^6 solar mass clusters.
